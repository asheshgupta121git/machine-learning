```python
import json
```

```python
# load data
def load_data(filename):
    with open('store_data.json', 'r') as f:
        data = json.load(f)

    return data
```

```python
data = load_data('store_data.json')

print(data)
print(type(data))
```

```python
# function to clean and structure the data
def clear_data(data):
    text_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    cleaned_data = []
    unique_users = set()
    
    for user in data:
        # clean rating
        raw_rating = str(user["rating"]).strip().lower()
        if(raw_rating in text_to_num):
            raw_rating = text_to_num[raw_rating]
        user["rating"] =  raw_rating
        
        #HANDLE MISSING VALUE
        raw_age = user.get("age")
        if(raw_age == None):
            user["age"] = None

        #duplication
        if(user["name"].strip() in  unique_users):
            continue 

        unique_users.add(user['name'])
        cleaned_data.append(user)

    return cleaned_data        
```

```python
clear_data(data)
```

```python
# Get meaningful insights form data
def get_insights(data):

    #avg rating
    tot_rating = 0
    for user in data:
        tot_rating += float(user["rating"])
    print(f"total rating = {tot_rating} avg rating = {tot_rating/len(data)}")

    # percentage of user with poor rating
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) < 3):
            poor_ratings += 1
    print(f"percentage of users with poor rating = {poor_ratings / len(data)*100}%")
```

```python
get_insights(data)
```

## Build a product recommendation feature (Rule based, not ML)
- If user rating >= 4, recommend same brand products.

- If user rating < 4, recommend different brand products.

```python
# recomendation feature. 
def get_recomendations(data):
    recomendation = []

    for user in data:
        curr_recomend = {}
        curr_recomend['name'] = user['name']

        if(float(user['rating']) >= 4):
            curr_recomend['brand'] = 'Apple'
        else:
            curr_recomend['brand'] = 'Samsung'

        recomendation.append(curr_recomend)
    return recomendation
```

```python
get_recomendations(data)
```
